```python
class MobileContact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# 📥 Sample Input
contact = MobileContact("Anita", "9876543210")

# 📤 Expected Output
contact.display_contact()
```
