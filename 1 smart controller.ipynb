```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Default status

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# 📥 Sample Input
light = SmartLight("Bedroom Light")
action = "ON"

if action == "ON":
    light.turn_on()
else:
    light.turn_off()

# 📤 Expected Output
light.display_status()
```
